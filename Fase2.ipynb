```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=1
b=0.002
c=1
d=0.001
x0=20
y0=20
t0=0
tf=30
h=0.01
E=0.5
```

```python
nt = int((tf-t0)/h) #número de pasos
Nx = np.empty(nt+1)
Nx[0]=x0
```

```python
nt = int((tf-t0)/h) #número de pasos
Ny = np.empty(nt+1)
Ny[0]=y0
```

```python
for step in range(nt):
  x1=a*Nx[step]-b*Nx[step]*Ny[step]-E*Nx[step]
  y1=-c*Ny[step]+d*Nx[step]*Ny[step]-E*Ny[step]
  x2=a*x1-b*(x1*Ny[step]+Nx[step]*y1)-E*x1
  y2=-c*y1+d*(x1*Ny[step]+Nx[step]*y1)-E*y1
  Nx[step+1] = Nx[step]+h*x1+0.5*h**2*x2
  Ny[step+1] = Ny[step]+h*y1+0.5*h**2*y2
```

```python
plt.plot(Nx)
```

```python
plt.plot(Ny)
```

```python
plt.plot(Nx,Ny)
```

Con odeint

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
def df_dt(x, t, a, b, c, d):
    """Función del sistema en forma canónica"""
    dx = a * x[0] - b * x[0] * x[1]-E*x[0]
    dy = - c * x[1] + d * x[0] * x[1]-E*x[1]
    return np.array([dx, dy])
conds_iniciales = np.array([x0, y0])
# Condiciones para integración
N = 3000
t = np.linspace(0, tf, N)
solucion = odeint(df_dt, conds_iniciales, t, args=(a, b, c, d))
```

```python
plt.plot(t, solucion[:, 0])
plt.plot(t, solucion[:, 1])
```

```python
plt.plot(solucion[:, 0], solucion[:, 1])
```
